```python
import pandas as pd

csv = pd.read_csv('./data/서울특별시_안전비상벨위치_20220316.csv',encoding='cp949')
```

```python
pd.set_option('display.max_row', 500)
```

```python
address = csv['위치']
```

```python
drop_ind = []
for ind, i in enumerate(address):
    if '~' in i:
        csv = csv.append({'name' : csv.iloc[ind]['name']+'_start', '위치' : i.split('~')[0]},ignore_index=True)
        plus = 4 - len(i.split('~')[1].split(' '))
        csv = csv.append({'name' : csv.iloc[ind]['name']+'_end', '위치' : " ".join(i.split('~')[0].split(' ')[:plus]) +' '+ i.split('~')[1]},ignore_index=True)
        drop_ind.append(ind)
```

```python
csv = csv.drop(drop_ind).reset_index(drop=True)
```

```python
from geopy.geocoders import Nominatim
geo_local = Nominatim(user_agent='South Korea')

# 위도, 경도 반환하는 함수
def geocoding(address):
    geo = geo_local.geocode(address)
    x_y = [geo.latitude, geo.longitude]
    return x_y
```

```python
address= csv['위치']
```

```python
#####주소를 위,경도 값으로 변환하기 #####
latitude = []
longitude =[]

for i in address:
    try:
        latitude.append(geocoding(i)[0])
        longitude.append(geocoding(i)[1])
    except:
        latitude.append(' ')
        longitude.append(' ')

csv['위도'] = latitude
csv['경도'] = longitude

csv
```

```python
# 카카오API를 사용하여 주소->좌표 변환
import requests, json

def get_location(address):
  url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
  # 'KaKaoAK '는 그대로 두시고 개인키만 지우고 입력해 주세요.
  # ex) KakaoAK 6af8d4826f0e56c54bc794fa8a294
  headers = {"Authorization": "KakaoAK 75f0db15d0b427fd9e0f1bc84724a452"}
  api_json = json.loads(str(requests.get(url,headers=headers).text))
  try:
    address = api_json['documents'][0]['address']
  except:
    return {"lat": ' ', "lng": ' '}
  crd = {"lat": str(address['y']), "lng": str(address['x'])}
  address_name = address['address_name']

  return crd
```
